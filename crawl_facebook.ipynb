```python
list_cmt=driver.find_elements(By.CSS_SELECTOR,"div.x1r8uery.x1iyjqo2.x6ikm8r.x10wlt62.x1pi30zi")
print(len(list_cmt))
```

```python
from bs4 import BeautifulSoup
cmt_full=[]
for i in list_cmt:
   html_content = i.get_attribute("outerHTML")
   soup = BeautifulSoup(html_content, 'html.parser')
   cmt=[]
   ten = soup.find('span', class_='x193iq5w')
   if ten:
      #print(ten.text.strip())
      cmt.append(ten.text.strip())
   comment = soup.find('div', class_='xdj266r x11i5rnm xat24cr x1mh8g0r x1vvkbs')
   if comment: 
      #print( comment.text.strip())   
      cmt.append(comment.text.strip())
   cam_xuc=soup.find('span',class_='xi81zsa x1nxh6w3 x1fcty0u x1sibtaa xexx8yu xg83lxy x18d9i69 x1h0ha7o xuxw1ft')   
   if cam_xuc:
      #print(cam_xuc.text.strip())
      cmt.append(cam_xuc.text.strip())
   cmt_full.append(cmt)   
print(len(cmt_full))   
for i in cmt_full:
   print(i)   
```

```python
import csv
csv_file_path = "du_lieu.csv"

# Mở tệp CSV để ghi (nếu tệp không tồn tại, nó sẽ được tạo)
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    # Viết dòng tiêu đề nếu cần
    # writer.writerow(['Column1', 'Column2', ...]) # Thay 'Column1', 'Column2', ... bằng tên các cột nếu cần

    # Viết dữ liệu từ danh sách cmt_full vào tệp CSV
    writer.writerows(cmt_full)
    
print("Dữ liệu đã được lưu vào", csv_file_path)
```
